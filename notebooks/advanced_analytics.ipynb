```python
# Rolling Sharpe Ratio Analysis
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load NAV data
nav = pd.read_csv("../data/raw/08_nav_history.csv")

# Create returns
pivot_nav = nav.pivot(
    index="date",
    columns="amfi_code",
    values="nav"
)

returns = pivot_nav.pct_change()

# Rolling Sharpe Ratio
rolling_sharpe = (
    returns.rolling(90).mean()
    /
    returns.rolling(90).std()
) * (252 ** 0.5)

# Plot
rolling_sharpe.plot(figsize=(12,5))
plt.title("Rolling Sharpe Ratio")
plt.show()

# Save chart
plt.savefig(
    "../reports/rolling_sharpe_chart.png",
    bbox_inches="tight"
)
```
